# Module 1 Simple Code Along

Goal: learn the Python basics we need before building the product catalog.

Core idea:

> Python code is values, decisions, loops, functions, and clear errors.

## 1. Values and variables

A variable gives a name to a value.

````python
name = "Keyboard"
price = 5499.0
in_stock = True

print(name)
print(price)
print(in_stock)
````

## 2. Basic types

The same product has strings, numbers, booleans, and lists.

````python
product_id = 1                 # int
name = "Keyboard"             # str
price = 5499.0                # float
in_stock = True               # bool
tags = ["keyboard", "mech"]  # list

print(type(product_id))
print(type(name))
print(type(price))
print(type(in_stock))
print(type(tags))
````

## 3. Simple calculations

Python can calculate and format values.

````python
price = 1000
tax_rate = 0.18

tax = price * tax_rate
total = price + tax

print(tax)
print(total)
print(f"Total: Rs {total}")
````

## 4. Lists

A list stores multiple values in order.

````python
product_names = ["Cable", "Keyboard", "Mouse"]

print(product_names[0])
print(product_names[1])

product_names.append("Speaker")
print(product_names)
````

## 5. Dicts

A dict stores named fields. This becomes very important for JSON and APIs.

````python
product = {
    "id": 1,
    "name": "Cable",
    "price": 499.0,
    "in_stock": True,
}

print(product["name"])
print(product["price"])

product["category"] = "Electronics"
print(product)
````

## 6. If / else

Use `if` when code needs to make a decision.

````python
price = 5499

if price < 1000:
    print("budget")
elif price < 5000:
    print("mid range")
else:
    print("premium")
````

## 7. Loops

Use a loop to do the same operation for every item.

````python
products = [
    {"name": "Cable", "price": 499},
    {"name": "Keyboard", "price": 5499},
    {"name": "Mouse", "price": 999},
]

for product in products:
    print(product["name"], product["price"])
````

## 8. Functions

A function names a reusable piece of logic.

````python
def price_label(name, price):
    return f"{name}: Rs {price}"


print(price_label("Cable", 499))
print(price_label("Keyboard", 5499))
````

## 9. Function parameters with defaults

Defaults make common calls shorter.

````python
def create_product(id, name, price, category="Misc"):
    return {
        "id": id,
        "name": name,
        "price": price,
        "category": category,
    }


print(create_product(1, "Cable", 499))
print(create_product(2, "Keyboard", 5499, category="Electronics"))
````

## 10. Type hints

Type hints make function intent clearer. Python does not enforce these by itself.

````python
def discount_price(price: float, percent: float) -> float:
    discount = price * percent / 100
    return price - discount


print(discount_price(1000, 10))
print(discount_price.__annotations__)
````

## 11. Exceptions

Use exceptions when something is wrong and the normal flow cannot continue.

````python
def validate_price(price):
    if price < 0:
        raise ValueError("price must be non-negative")
    return price


print(validate_price(499))

# Uncomment this line to see the exception:
# validate_price(-50)
````

## 12. Catching exceptions

Catch exceptions when you can handle them usefully.

````python
try:
    validate_price(-50)
except ValueError as error:
    print("Could not create product:", error)
````

## 13. A small Product class

This is the bridge into the product catalog project.

````python
class Product:
    def __init__(self, id: int, name: str, price: float):
        if price < 0:
            raise ValueError("price must be non-negative")
        self.id = id
        self.name = name
        self.price = price

    def to_dict(self):
        return {
            "id": self.id,
            "name": self.name,
            "price": self.price,
        }


product = Product(1, "Cable", 499)

print(product.name)
print(product.to_dict())
````
